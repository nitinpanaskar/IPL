### 2. Data loading and description <a class ="anchor" id="section2"></a>


- The data set contains the details of the Teams that have participated in IPL Till 2018. 
- The dataset comprises of __696 observations of 18__ columns
- Below is a table showing names of all the columns and their description.

| Column Name   | Description                                               |
| ------------- |:-------------                                            :| 
| Id            | Identification Number                                     | 
| Season        | IPL Year                                                  |  
| City          | City where IPL Match was played                           | 
| Date          | IPL Match Date                                            |   
| Team1         | Team 1 Participating in IPL                               |

|umpire2        | Umpire Name-2                                             



About IPL: The Indian Premier League (IPL) is a professional Twenty20 cricket league in India contested during April and May of every year by teams representing 8 Indian cities and some states.The league was founded by the Board of Control for Cricket in India (BCCI) in 2008.The IPL is the most-attended cricket league in the world and in 2014 ranked sixth by average attendance among all sports leagues.In 2010, IPL became the first sporting event in the world to be broadcast live on YouTube.

The brand value of IPL in 2018 US Dollar 6.3 billion, according to Duff & Phelps. According to BCCI, the 2015 IPL season contributed ₹11.5 billion (US$182 million) to the GDP of the Indian economy.

### 5.12 Most Match Winners <a class ="anchor" id="section21"></a>

```python
#know top 10 players
top_players = matches.player_of_match.value_counts()[:10]
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Most Match Winners")
#top_players.plot.bar()
sbn.barplot(x = top_players.index, y = top_players, orient='v', palette="Set2");
plt.xticks (rotation=90)
plt.show()
```

### 5.14 Which team defends the score?<a class ="anchor" id="section23"></a>

```python
'''Which team defends the score?'''
matches[matches['win_by_runs']>0].groupby(['winner'])['win_by_runs'].apply(np.median).sort_values(ascending = False)
```

```python
fig, ax = plt.subplots()
ax.set_title("Winning by Runs - Team Performance")
sbn.boxplot(y = 'winner', x = 'win_by_runs', data=matches[matches['win_by_runs']>0], orient = 'h'); #palette="Blues");
plt.show()
```

### 5.15 Which team chase the score?<a class ="anchor" id="section24"></a>
